```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
data = {
    "Transaction_ID":[1,2,3,4,5,6,7,8],
    "Product1":["Milk","Milk","Bread","Milk","Tea","Milk","Bread","Tea"],
    "Product2":["Bread","Butter","Butter","Bread","Sugar","Eggs","Jam","Milk"],
    "Product3":["Butter",None,None,"Eggs",None,None,None,"Sugar"]
}
data=pd.DataFrame(data)
data
```

```python
transactions=data.iloc[:,1:].values.tolist()
transactions
```

```python
new_transactions = []
for row in transactions:
    temp = []
    for item in row:
        if pd.notnull(item):
            temp.append(item)
    new_transactions.append(temp)

transactions = new_transactions
transactions
```

```python

te = TransactionEncoder()
transactions = te.fit(transactions).transform(transactions)

transactions = pd.DataFrame(transactions, columns=te.columns_)

transactions
```

```python
frequent_items=apriori(transactions, min_support=0.2)
print(frequent_items)
```

```python
rules=association_rules(frequent_items)
rules
```
