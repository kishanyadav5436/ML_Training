### Step 20: Remove Outliers

```python
def remove_outliers(df,column):

    Q1 = df[column].quantile(0.25)

    Q3 = df[column].quantile(0.75)

    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR

    upper = Q3 + 1.5 * IQR

    return df[(df[column]>=lower) & (df[column]<=upper)]
```

```python
# Remove outliers only from feature columns
for col in numerical_cols:
    if col != 'Annual_Salary_LPA':
        df = remove_outliers(df, col)
```

### Step 23: Feature Encoding

##### Label Encoding

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df['Gender']=le.fit_transform(df['Gender'])

df['Remote_Work']=le.fit_transform(df['Remote_Work'])
```

```python
job_mapping = {'Junior':1,'Mid':2,'Senior':3,'Lead':4,'Manager':5}

df['Job_Level'] = df['Job_Level'].map(job_mapping)
```

##### One Hot Encoding

```python
# df=pd.get_dummies(df,columns=['Education','Department','City'],drop_first=True,dtype=int)
```

```python
df['Education']=df['Education'].map({'Diploma':0, 'Bachelor':1, 'Master':2,'PhD':3})
df['Department']=df['Department'].map({'Operations':0, 'IT':1, 'Finance':2,'Sales':3,'HR':4,'Marketing':5})
df['City']=df['City'].map({'Hyderabad':0, 'Mumbai':1, 'Pune':2,'Chennai':3,'Bangalore':4, 'Delhi':5})
```

### Step 24: Separate Features & Target

```python
X = df.drop('Annual_Salary_LPA', axis=1)

y = df['Annual_Salary_LPA']
```

### Step 25: Train-Test Split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.20,random_state=42)
```

### Step 26: Feature Scaling

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)

X_test = scaler.transform(X_test)
```

### Step 27: Train Multiple Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from xgboost import XGBRegressor

lr = LinearRegression()
rf = RandomForestRegressor()
gbr = GradientBoostingRegressor()
xg = XGBRegressor()

lr.fit(X_train, y_train)
rf.fit(X_train, y_train)
gbr.fit(X_train, y_train)
xg.fit(X_train, y_train)
```

### Step 28: Prediction

```python
y_pred1 = lr.predict(X_test)
y_pred2 = rf.predict(X_test)
y_pred3 = gbr.predict(X_test)
y_pred4 = xg.predict(X_test)
```

### Step 29: Model Evaluation

```python
from sklearn.metrics import (mean_absolute_error,mean_squared_error,r2_score)
import numpy as np

score1 = r2_score(y_test, y_pred1)
score2 = r2_score(y_test,y_pred2)
score3 = r2_score(y_test,y_pred3)
score4 = r2_score(y_test,y_pred4)

print("R² Score :", score1)
print("R² Score :", score2)
print("R² Score :", score3)
print("R² Score :", score4)
```
