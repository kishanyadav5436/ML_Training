## np.arange()

```python
arr = np.arange(1,11,2)
print(arr)
print(arr.ndim)
```

```python
arr = np.arange(1,10)
print(arr)
print(arr.shape)
print(arr.reshape(3,3))
arr = arr.reshape(3,3)
print(arr.ndim)
```
