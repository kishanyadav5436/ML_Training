#### Array Dimentions

```python
## 0 Dimentional Array
import numpy as np
arr1 = np.array(50)
print(arr1)
print(type(arr1))
print(arr1.shape)
print(arr1.ndim)
```

```python
## 1 Dimentional Array
import numpy as np
arr1 = np.array([50,89,56,45])
print(arr1)
print(type(arr1))
print(arr1.shape)
print(arr1.ndim)
```

```python
## 2 Dimentional Array
import numpy as np
arr1 = np.array([[50,89],[56,45],[56,45]])
print(arr1)
print(type(arr1))
print(arr1.shape)
print(arr1.ndim)
```

```python
## 3 Dimentional Array
import numpy as np
arr1 = np.array([[[50,89],[56,45]],[[50,89],[56,45]]])
print(arr1)
print(type(arr1))
print(arr1.shape)
print(arr1.ndim)
```

```python
import numpy as np
arr1 = np.array([[12,18],[16,20]])
arr2 = np.array([[2,3],[4,5]])
print(arr1)
print(arr2)
print(arr1*arr2)
```

```python
import numpy as np
arr1 = np.array([[12,18],[16,20]])
arr2 = np.array([[2,3],[4,5]])
print(arr1@arr2)
```
