```python
import pandas as pd

# Create DataFrame
df = {
    "Student": ["A", "B", "C", "D", "E"],
    "Maths_Rank": [1, 2, 3, 4, 5],
    "Physics_Rank": [2, 3, 1, 4, 5]
}

df = pd.DataFrame(df)
print(df)
```

```python
import matplotlib.pyplot as plt

import seaborn as sns

plt.figure(figsize=(8,5))

sns.lineplot(data=df, x="Student", y="Maths_Rank", marker="o", label="Maths Rank")

sns.lineplot(data=df,x="Student",y="Physics_Rank",marker="o", label="Physics Rank" )

plt.title("Maths Rank vs Physics Rank")

plt.ylabel("Rank")

plt.gca().invert_yaxis()

#Rank 1 appo

plt.grid(True)

plt.show()
```
