```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.array([10,20,30,40])
y=np.array([100,200,300,400])
plt.plot(x,y, marker="*")
plt.show()
```

```python
products=['Mobile','pendrive','laptop','watch']
sales=[30000,20000,4000,8000]
plt.bar(products,sales,width=0.5, color='red')
plt.show()
```

```python
products=['Mobile','pendrive','laptop','watch']
sales=[30000,20000,40000,80000]
plt.barh(products,sales,height=0.5, color='blue')

plt.xlabel("Products")
plt.ylabel("Sales")
plt.title("Product Sales")
plt.show()
```

```python
products=['phone', 'laptop','binders','copiers']
prices=[34000,25000,38000,40000]
plt.pie(prices, labels=products)
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

data = pd.read_excel('Superstore.xlsx')
data.head()
```

```python
data.columns
```

```python
category_sales = data.groupby('Category')['Sales'].sum()
print(category_sales)
plt.figure(figsize=(4,3))
category_sales.plot(kind='bar')
plt.xticks(rotation=45)
plt.show()
```

```python

region_wise_profit = data.groupby('Region')['Profit'].sum()

plt.figure(figsize=(4,3))
region_wise_profit.plot(kind='barh')
plt.title("Region-wise Profit")
plt.xlabel("Profit")
plt.ylabel("Region")
plt.show()
```

```python
category_sales = data.groupby('Category')['Sales'].sum()
plt.pie(category_sales, labels=category_sales.index, autopct='%1.1f%%', startangle=180)

# plt.figure(figsize=(4,3))
# category_sales.plot(kind='pie')
# plt.xticks(rotation=45)
plt.show()
```
