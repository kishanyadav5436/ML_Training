## 13. Students with Marks Between 90 and 100

```
df[(df["Marks"]>=90) & (df["Marks"]<=100)]

# Count
print(df[(df["Marks"]>=90) & (df["Marks"]<=100)].shape[0])
```

## 14. Using between()

```
df[df["Marks"].between(90,100)]

# Count
print(df[df["Marks"].between(90,100)].shape[0])
```

## 15. Average Marks

```
df["Marks"].mean()
```

## 16. Apply Method

```
df["Half_Marks"] = df["Marks"].apply(lambda x: x/2)
df

# Male/Female
df["Male_Female"] = df["Gender"].apply(lambda x: "1" if x=="Male" else "0")
df
```

## 17. Map Function

```
gender_map={"Male":1,"Female":0}
df["Gender_Code"]=df["Gender"].map(gender_map)
df
```

## 18. Drop Columns

```
df.drop(["Half_Marks","Male_Female"], axis=1, inplace=True)
df
```
