```python
data['Bathrooms'].describe()
```

```python
data.fillna({'Bathrooms': data['Bathrooms'].mean()}, inplace=True)
```

```python
round(data.isnull().sum().sort_values(ascending=False)/len(data),2)*100
```

```python
data.fillna({'Balconies': data['Balconies'].mean()}, inplace=True)
```

```python
data['Parking_Available'].describe()
```

```python
print(data["Bedrooms"].unique())
```

```python
mapping = {
    "One": 1,
    "Two": 2,
    "Three": 3,
    "Four": 4,
    "Five": 5,
    "Six": 6
}

data["Bedrooms"] = data["Bedrooms"].replace(mapping)
```

```python
data["Bedrooms"] = pd.to_numeric(data["Bedrooms"], errors="coerce")
```

```python
data["Bedrooms"].dtype
```

```python
data["Bedrooms"] = data["Bedrooms"].fillna(data["Bedrooms"].median())
```

```python
mapping = {
    "One": 1,
    "Two": 2,
    "Three": 3,
    "Four": 4,
    "Five": 5
}

data["BHK"] = data["BHK"].replace(mapping)

data["BHK"] = data["BHK"].astype(str).str.replace("BHK", "", regex=False)

data["BHK"] = pd.to_numeric(data["BHK"], errors="coerce")

data["BHK"] = data["BHK"].fillna(data["BHK"].median())
```

```python
data["City"] = data["City"].fillna(data["City"].mode()[0])
data["Area_Sqft"] = data["Area_Sqft"].fillna(data["Area_Sqft"].median())
data["Price_INR"] = data["Price_INR"].fillna(data["Price_INR"].median())
data["Bedrooms"] = data["Bedrooms"].fillna(data["Bedrooms"].median())
```

```python
print(data["Bedrooms"].unique())
print(data["BHK"].unique())
```

```python
print(data["Property_Age"].unique())
```

```python
data["Property_Age"] = data["Property_Age"].replace({
    "New": 0
})
```

```python
data["Property_Age"] = pd.to_numeric(data["Property_Age"], errors="coerce")
```

```python
data["Property_Age"] = data["Property_Age"].fillna(data["Property_Age"].median())
```

```python
round(data.isnull().sum().sort_values(ascending=False)/len(data),2)*100
```

```python
print(data["Parking_Available"].unique())
```

```python
data["Parking_Available"] = (
    data["Parking_Available"]
    .astype(str)
    .str.strip()
    .str.upper()
)
```

```python
print(data["Parking_Available"].unique())
```

```python
import numpy as np

data["Parking_Available"] = data["Parking_Available"].replace("NAN", np.nan)
```

```python
data["Parking_Available"] = data["Parking_Available"].fillna(
    data["Parking_Available"].mode()[0]
)
```

```python
data["Total_Floors"] = pd.to_numeric(
    data["Total_Floors"], errors="coerce"
)

data["Total_Floors"] = data["Total_Floors"].fillna(
    data["Total_Floors"].median()
)
```

```python
round(data.isnull().sum() / len(data) * 100, 2)
```
